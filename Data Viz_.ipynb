```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV Data
df = pd.read_csv("C:/Users/obafe/Downloads/Top 50 US Tech Companies 2022 - 2023.csv")
```

```python
new_df=df.drop(['Industry'],axis=1)
```

```python

# Prepare Data for Plotting
categories = df['HQ State']
values = df["Market Cap (USD in Trillions)"]

# Create the Bar Graph
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph from Top 50 Tech Companies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ax=sns.boxplot(x=new_df["Market Cap (USD in Trillions)"])
```

```python
# calculate correlation matrix
corr = new_df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```
